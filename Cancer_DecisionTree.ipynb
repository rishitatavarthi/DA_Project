```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Coimbra_breast_cancer_dataset.csv')
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
target = df["Classification"]
print(target)
```

```python
feat_list=['Glucose','Insulin','Resistin']
feat=df[feat_list]
print(feat)
```

```python
model=DecisionTreeClassifier()
model.fit(feat,target)

```

```python
DT=model.fit(feat,target)
```

```python
glu=int(input("Enter Glucose: "))
insu=int(input("Enter Insulin: "))
res=int(input("Enter Resistin: "))
pred=model.predict([[glu,insu,res]])
print(pred)
```

```python
if pred==1:
  print("No Coimbra breast cancer")
else:
  print(" Coimbra breast cancer")
```
