```python
import pandas as pd
```

```python
df=pd.read_csv('/content/diabetcsv (1).csv')
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
target=df["class"]
print(target)
```

```python
feat_list=['insu','mass','age','skin','plas']
feat=df[feat_list]
print(feat)
```

```python
model=RandomForestClassifier()
model.fit(feat,target)
```

```python
insu=int(input("Enter Insulin: "))
mass=int(input("Enter Mass: "))
age=int(input("Enter Age: "))
plas=int(input("Enter Plas: "))
skin=int(input("Enter Skin: "))
pred=model.predict([[insu,mass,age,plas,skin]])
print(pred)
```

```python
if pred == ['tested_positive']:
  print("You have diabetes, Maintain Proper Medication")
else:
  print("You can freely enjoy eating Sweets")
```
